```python
scores = np.random.uniform(50, 100, 100)
plt.hist(scores, bins=10, edgecolor='black')
plt.title("Question 5")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.grid(True, linestyle='--', alpha=0.7) 

plt.annotate('Annotation', xy=(3, 1), xytext=(3, 2),
             arrowprops=dict(facecolor='black', shrink=0.05))
```

```python
scores_1 = np.random.uniform(60, 100, 50)  
scores_2 = np.random.uniform(60, 100, 50)  

plt.boxplot([scores_1, scores_2], patch_artist=True)

plt.title("Question 6")
plt.xlabel("Class")
plt.ylabel("Scores")
plt.xticks([1, 2], ['Class 1', 'Class 2'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
salespeople = ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve']
product_A = [30, 25, 35, 20, 30]
product_B = [20, 30, 15, 25, 20]
product_C = [25, 15, 20, 30, 25]

x = np.arange(len(salespeople))

fig, ax = plt.subplots()

bars_A = ax.bar(x, product_A, label='Product A', color='skyblue')
bars_B = ax.bar(x, product_B, bottom=product_A, label='Product B', color='lightgreen')
bars_C = ax.bar(x, product_C, bottom=np.array(product_A) + np.array(product_B), label='Product C', color='salmon')

for bars in [bars_A, bars_B, bars_C]:
    for bar in bars:
        yval = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_y() + yval / 2, str(int(yval)),
                ha='center', va='center', color='black')

ax.set_title('Total Sales by Salesperson and Product')
ax.set_xlabel('Salespeople')
ax.set_ylabel('Number of Products Sold')

ax.set_xticks(x)
ax.set_xticklabels(salespeople)
ax.legend()


```
